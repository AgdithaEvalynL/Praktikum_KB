## membaca data
### menggunakan read_csv agar tertampil

```python
df = pd.read_csv("filmtv_movies - ENG.csv") 
```

## membuat dictionary
#### membuat key dengan ketentuan postest dan value nya menggunakan syntax seperti dibawah ini


```python
##dict

info = {
    'JumlahRecord': df.shape[0],
        'NamaAtribut': df.columns.values,
        'JumlahAtribut(objek)': df.select_dtypes(['object']).columns,
        'JumlahAtribut(Angka)':df.select_dtypes(['float64','int64']).columns,
        'JumlahAtribut(null)':df.loc[:, df.isnull().any()].columns}
        
```

## menampilkan informasi

```python
print(" Jumlah Record = ", info['JumlahRecord'])
print("\n Nama Atribut = ", info['NamaAtribut'])
print("\n Nama Atribut(objek) = ", info['JumlahAtribut(objek)'])
print("\n Jumlah Atribut(objek) = ", len(info['JumlahAtribut(objek)']))
print("\n Nama Atribut(Angka) = ", info['JumlahAtribut(Angka)'])
print("\n Jumlah Atribut(Angka) = ", len(info['JumlahAtribut(Angka)']))
print("\n Nama Atribut(null) = ", info['JumlahAtribut(null)'])
print("\n Jumlah Atribut(null) = ", len(info['JumlahAtribut(null)']))
```
