```python
import pandas as pd
import numpy as np 
from matplotlib import pyplot as plt
import seaborn as sns 
```

## membaca file csv dengan library pd

```python
df = pd.read_csv("filmtv_movies - ENG.csv") 
```

```python
df["year"] = pd.to_datetime(df["year"])
```

#### dalam grafik dibawah ini kita bisa mendapatkan informasi bahwa banyak public vote yang di dapatkan di semua genre dengan rata-rata 1.0 dengan perbedaan public_vote. Dan didapatkan public vote terbanyak adalah di 7.0 dan di countnya terhitung 3.0 .  

```python
# Dengan seaborn
dataplotting=df.sort_values("title").head(30)
sns.set_style("ticks")
plt.figure(figsize = (8,9))
sns.countplot(data = dataplotting, x = "genre", hue = "public_vote", palette = "rocket")
plt.title('Perbandingan genre dengan banyak nya public vote ')
```

### didapatkan informasi dari grafik dibawah adalah menjelaskan hubungan korelasi antara 2 attribut yang tipe data nya numeric. 

```python
plt.figure(figsize=(10,5))

#Argumen wajib corr(korelasi) | Cari seaborn pallete
sns.heatmap(
    df.corr(),
    annot=True,
    cmap='flare'
) 
plt.title('korelasi antar attribut variabel ')

plt.show()
```

### menggambar diagram pie dengan attribut humor 

```python
data_pie = df["humor"].value_counts()
data_pie.head()
```

```python
data_pie = df["humor"].value_counts().rename_axis("humor")
data_pie.head()
```

```python
data_pie = df["humor"].value_counts().rename_axis("humor").reset_index(name = "id_count")
```

### informasi dari data pie ini bisa kita liat seperti tabel di atas, bahwa seluruh title film yang mengandung dengan nilai 0 (humor) ada 23591. diagram pie memudahkan kita cek menggunakan dalam bentuk persen  

```python
plt.pie(data_pie.id_count,
    labels = data_pie["id_count"], 
    startangle=90, 
    autopct='%.1f%%', 
    shadow=True
)
plt.title('diagram pie perbandingan humor dalam sebuah film')
plt.show()
```

### grafik dibawah ini menunjukkan sebuah implementasi data yang dimana kita bisa liat sebuah title film berada di genre apa berdasarkan dot/titik dibawah. sebagai contoh $5 day berada di genre comedy.
